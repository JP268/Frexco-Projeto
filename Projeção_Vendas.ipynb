# Importando os dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('Dados.xlsx')
display(df)
```

# Total de Vendas

```python
total_vendas = df['Vendas'].sum()
total_vendas
```

# Criando um gr√°fico dos dados

```python
plt.figure(figsize = (15,6))

plt.plot(df['Data'], df['Vendas'])
plt.show
```

# Separando cada ano

```python
# Transformando a coluna data em String
df['Data'] = df['Data'].astype(str)

# Criando uma tabela apenas o ano de 2022
df_2022 = df.loc[df['Data'].str.contains('2022')]


# Criando uma tabela apenas o ano de 2023
df_2023 = df.loc[df['Data'].str.contains('2023')]

```

# Criando gr√°fico do ano de 2022 e o total de vendas

```python
plt.figure(figsize = (15,6))

plt.plot(df_2022['Data'], df_2022['Vendas'])
plt.show()

total_vendas_2022 = df_2022['Vendas'].sum()
total_vendas_2022
```

# Criando gr√°fico do ano de 2023 e o total de vendas

```python
plt.figure(figsize = (15,6))

plt.plot(df_2023['Data'], df_2023['Vendas'])
plt.show()

total_vendas_2023 = df_2023['Vendas'].sum()
total_vendas_2023
```

```python
plt.figure(figsize = (15,6))

plt.plot(df_2022['Data'], df_2022['Vendas'], label = '2022')
plt.plot(df_2023['Data'], df_2023['Vendas'], label = '2023')


plt.legend()
plt.show()
```

### Realizando a porcentagem de aumento que teve de um m√™s para o outro, e vendo quanto vai ser a demando necess√°ria para os 5 dias

```python
# Total de vendas do dia 6 at√© o dia 20 - 12/2022
df_2022[0:15]['Vendas'].sum()
```

```python
# Total de vendas do dia 6 at√© o dia 20 - 01/2023
df_2023[5:20]['Vendas'].sum()
```

```python
# Porcentagem de aumento 
pa = (df_2023[5:20]['Vendas'].sum() * 100) / df_2022[0:15]['Vendas'].sum()

#Arredondando a porcentagem
round(pa)
```

```python
# Total de vendas do dia 21 at√© o dia 25 - 12/2022
df_tt22 = df_2022[15:20]['Vendas'].sum()
df_tt22
```

```python
# Previs√£o de vendas do dia 21 at√© o dia 25 - 01/2023
x = (df_2022[15:20]['Vendas'].sum() * pa) / 100

#Arredondando o valor
round(x)
```
